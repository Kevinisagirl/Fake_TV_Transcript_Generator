```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://transcripts.wikia.com/wiki/Adventure_Time"
url_base = "http://transcripts.wikia.com"
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
lsts = soup.find("div", {"id": "mw-content-text"}).find_all("ol")
transcript_links = []
for ol in lsts:
    a = (ol.find_all("a"))
    for link in a:
        transcript_links.append(url_base + link['href'])
len(transcript_links)
```

```python
character_lines_df = pd.DataFrame(columns=['Character', 'Line', 'Episode'])
for link in transcript_links:
    try:
        html = urlopen(link)
    except:
        link = url_base + link.split("//")[2]
        html = urlopen(link)
    soup = BeautifulSoup(html)
    transcript = soup.find("div", {"id": "mw-content-text"}).find_all('li')
    # parse the transcipt into lines
    line = ["Action: " + lst.get_text(strip=False)[0:-1] if lst.get_text(strip=False)[0:-1].startswith('[') else lst.get_text(strip=False)[0:-1] for lst in transcript]
    line = [string.replace(u'\xa0', u' ') for string in line]
    cleaned_line = []
    # we want to split the lines into two lists
    # the first element is the character
    # the second is the line
    # however, sometimes there are ':' in the lines
    # so we have to rejoin the later element
    for string in line:
        string_list = string.split(':')
        new_str_lst = []
        new_str_lst.append(string_list.pop(0))
        new_str_lst.append("".join(string_list))
        cleaned_line.append(new_str_lst)
    # get the episode title
    episode_title = str(soup.title.text).split(' | ')[0]
    # make into a dataframe
    try:
        df = pd.DataFrame(cleaned_line, columns=['Character', 'Line'])
        df['Episode'] = episode_title
    except:
        print(len(l) for l in cleaned_line)
        print(episode_title)
        break
    # add to the main df
    character_lines_df = character_lines_df.append(df, ignore_index=True)
character_lines_df.head()
```

```python
character_lines_df.to_csv("AT_lines.csv", index=False)
character_lines_df[character_lines_df.Character != "Action"].to_csv("AT_lines_noActions.csv", sep="@", index=False)
set(character_lines_df.Character.tolist())
```

```python
# LSP word cloud!!
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# make shape mask
lsp_mask = np.array(Image.open("Lsp_2.png"))
lsp_mask
text = " ".join(lsp_df.Line)
```

```python
def make_wc(mask, text, cont_color, bkg,cw):
    # create coloring from image
    image_colors = ImageColorGenerator(mask)

    # Create and generate a word cloud image:
    wordcloud = WordCloud(max_font_size=70, max_words=150, 
                          mask = mask, background_color=bkg,
                         contour_width=cw, contour_color=cont_color).generate(text)

    # Display the generated image:
    plt.figure(figsize=[7,7])
    plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation='bilinear')
    plt.axis("off")
    plt.show()
text = " ".join(lsp_df.Line)
make_wc(lsp_mask,text,'violet', "white",3)
```
